```python
def bigram_by_step(lst, step= 1):
  """
  fow window size of 1, give step = 2
  """
  bigram_2 = []
  bigram_2_tags = []
  step_size = []
  for j in np.arange(step)[1:]:
    for i in tqdm.tqdm(range(len(lst)-step)):
        bigram_2.append((lst[i][0], lst[i+step-j][0] ))
        bigram_2_tags.append( [lst[i][1], lst[i+step-j][1]])
        step_size.append(step-j)

  bigram_df = pd.concat([pd.DataFrame(bigram_2, columns = ["w1","w2"]), pd.DataFrame(bigram_2_tags, columns = ["w1_t", "w2_t"]), pd.DataFrame(step_size, columns = ["length"])], axis = 1)
  return bigram_df
```

```python
def preprocess_bigrams(df,stop_words, df_lemm):
  freq_dict = (df_lemm[0].value_counts()).to_dict()
  bigram_freq = []
  df_new = df[((df["w1_t"] == "NOUN") & (df["w2_t"]== "NOUN")) | ( (df["w1_t"] == "ADJ") & (df["w2_t"]== "NOUN"))]
  for word in stop_words:
    df_new = df_new[(df_new["w1"] != word) & (df_new["w2"] != word) ]
  df_new_new = df_new.reset_index()
  df_new_new = df_new_new.drop(columns = ["index"])
  for i in tqdm.tqdm(range(df_new_new.shape[0])):
   

    if ((df_new_new[df_new_new.index == i].iloc[:, 0]).values[0].isalpha() and (df_new_new[df_new_new.index == i].iloc[:, 1]).values[0].isalpha()):
      pass
    else:
      df_new_new = df_new_new.drop(i)

  
  counts = pd.DataFrame(df_new_new.iloc[:,:2].value_counts())
  counts= counts.reset_index()
  counts.columns = ["w1","w2","c(w1w2)"]
  counts_new = counts[counts["c(w1w2)"] >= 10]
  last_colloc = counts_new.iloc[:,:3]

  for i in tqdm.tqdm(range(last_colloc.shape[0])):
    bigram_freq.append((freq_dict[last_colloc[last_colloc.index == i].iloc[:, 0].values[0]],freq_dict[last_colloc[last_colloc.index == i].iloc[:, 1].values[0]]))

  last_colloc = pd.concat([last_colloc, pd.DataFrame(bigram_freq, columns = ["c(w1)", "c(w2)"])], axis=1)

  return last_colloc
```

```python
def t_test(df, df_all, step = 1):
  if step !=3:
    null_hyp_w1 = df["c(w1)"]/len(df_all)
    null_hyp_w2 = df["c(w2)"]/len(df_all)
    null = null_hyp_w1.mul(null_hyp_w2.values)    

    p = df["c(w1w2)"]/len(df_all)
    mean, var = bernoulli.stats(p)
    t_score = -(null-mean ) / (np.sqrt(var/(len(df_all))))  
  else:
    null_hyp_w1 = df["c(w1)"]/(len(df_all))
    null_hyp_w2 = df["c(w2)"]/(len(df_all))
    null = null_hyp_w1.mul(null_hyp_w2.values)    

    p = df["c(w1w2)"]/(3*len(df_all))
    mean, var = bernoulli.stats(p)
    t_score = -(null-mean ) / (np.sqrt(var/(3*len(df_all))))  
  t_score = pd.DataFrame(t_score)
  t_score.columns = ["t_score"]
  df =pd.concat([df,t_score],axis=1 )

  return df                 
```

```python
def chi_square_test(bigram, bigram_all):

  o11 = bigram["c(w1w2)"] / len(bigram_all)
  o22 = (len(bigram_all) - bigram["c(w1w2)"]) / len(bigram_all)
  o12 = (bigram["c(w2)"] - bigram["c(w1w2)"]) / len(bigram_all)
  o21 = (bigram["c(w1)"] - bigram["c(w1w2)"]) / len(bigram_all)
  X2 = pd.DataFrame((len(bigram_all) * (o11*o22 - o12*o21)**2) / ((o11 + o12) * (o11 + o21) * (o12 + o22) * (o21 + o22)))
  X2.columns = ["chi-square_score"]
  bigram = pd.concat([bigram, X2], axis=1)
  return bigram   
```

```python
def likelihood_test(bigram, bigram_all):
  scores = []
  for i in range(len(bigram)):

    p= bigram["c(w2)"].iloc[i] / len(bigram_all)
    b_1 = binom.pmf(bigram["c(w1w2)"].iloc[i], bigram["c(w1)"].iloc[i], p)
    b_not_1 = binom.pmf((bigram["c(w2)"].iloc[i] - bigram["c(w1w2)"].iloc[i]), (len(bigram_all)- bigram["c(w1)"].iloc[i]), p)
    L_H1 = b_1 * b_not_1
    
    p_1 = bigram["c(w1w2)"].iloc[i] / bigram["c(w1)"].iloc[i]
    P_2 = (bigram["c(w2)"].iloc[i] - bigram["c(w1w2)"].iloc[i]) / (len(bigram_all)- bigram["c(w1)"].iloc[i])
    b_1_dep = binom.pmf(bigram["c(w1w2)"].iloc[i], bigram["c(w1)"].iloc[i], p_1)
    b_not_1_dep = binom.pmf((bigram["c(w2)"].iloc[i] - bigram["c(w1w2)"].iloc[i]), (len(bigram_all)- bigram["c(w1)"].iloc[i]), P_2)
    L_H2 = b_1_dep * b_not_1_dep
    if L_H1 == 0:
      L_H1 = 1e-40
    score = -2* np.log(L_H1/ L_H2)

    scores.append(score)

  scores = pd.DataFrame(scores, columns = [r"-2log(λ)"])
  bigram = pd.concat([bigram, scores], axis = 1)
  return bigram

```
